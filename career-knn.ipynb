```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
career_labels = [
    "Engineering",
    "Communications",
    "Healthcare",
    "Law",
    "Education",
    "Business",
    "Government",
    "Arts",
    "Sales",
    "Architecture"
]
label_encoder = LabelEncoder()

encoded_career_labels = label_encoder.fit_transform(career_labels)

print("Encoded Labels:", encoded_career_labels)
print("Label Mapping:")
for label, career in enumerate(label_encoder.classes_):
    print(f"Label {label}: {career}")
```

```python
label_to_career = {
    "Architecture": 0,
    "Arts": 1,
    "Business": 2,
    "Communications": 3,
    "Education": 4,
    "Engineering": 5,
    "Government": 6,
    "Healthcare": 7,
    "Law": 8,
    "Sales": 9
}

df['career'] = df['career'].map(label_to_career)

df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
df = df.drop(['personality_Extroverted'], axis=1)
df = df.drop(['thinking_Feeler'], axis=1)
df = df.drop(['perception_Judger'], axis=1)
df = df.drop(['thought_Intuition'], axis=1)
df.head()
```

```python
X = df.drop('career',axis=1).values
y = df['career'].values
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbors = np.arange(1,9)
for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
print(X)
```

```python
new_data = pd.DataFrame({
    'LRscore': 42,
    'personality_Introverted': [True],
    'thinking_Thinker': [False],
    'perception_Perceiver': [True],
    'thought_Sensing': [True]
})

predictions = knn.predict(new_data)

print("Predicted Career Label:", predictions[0])
```

```python
import joblib
joblib.dump(knn, 'careermodel.pkl')
```
